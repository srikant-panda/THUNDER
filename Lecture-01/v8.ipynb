```python
def get_data(raw_data : str):
    is_valid = validate_data(raw_data)
    if not is_valid:
        raise ValueError('Invalid syntax!')
    data = get_context(raw_data)
    if not data:
        raise ValueError('Invalid data!')
    # print(data)
    return data

def validate_data(raw_data : str):
    if not raw_data:
        return False
    if not 'console.log' in raw_data:
        return False
    return True

def get_context(raw_data):
    raw_data = raw_data.strip()
    if not raw_data:
        return
    data1 = raw_data.split('console.log(')
    final_data = data1[1][:-1]
    return final_data

class Operation:
    def __init__(self,raw_data):
        self.operands = []
        self.operatior : str = ''
        raw_data = get_data(raw_data)
        # print(raw_data)
        for i in ['+','-','*','/','%']:
            if i in raw_data:
                self.operands = raw_data.split(i)
                self.operatior = i
                break
            else:
                self.operands = [raw_data]
        return None
    def perform_operation(self):
        operaion = self.operatior
        try:
            match operaion:
                case '+':
                    try:
                        # return int(self.operands[0]) + int(self.operands[1])
                        result = 0
                        for i in self.operands:
                            result+=int(i)
                        return result
                    except ValueError:
                        return self.operands[0] + self.operands[1]
                case '-':
                    try:
                        return int(self.operands[0]) - int(self.operands[1])
                    except ValueError:
                        return self.operands[0] - self.operands[1]
                case '*':
                    try:
                        return int(self.operands[0]) * int(self.operands[1])
                    except ValueError:
                        raise ValueError('str cannot perform thi operation.')
                case '/':
                    try:
                        return int(self.operands[0]) / int(self.operands[1])
                    except ValueError:
                        raise ValueError('str cannot perform thi operation.')
                case '%':
                    try:
                        return int(self.operands[0]) % int(self.operands[1])
                    except ValueError:
                        raise ValueError('str cannot perform thi operation.')           
                case _:
                    if self.operands[0].startswith('\'') and self.operands[0].endswith('\''):
                        return self.operands[0][1:-1]
                    return self.operands[0]
        except IndexError:
            pass
                
def main(raw_data):
    o1 = Operation(raw_data)
    print(o1.perform_operation())

if __name__ == "__main__":
    with open('sample.js','r') as fp:
        data = fp.readlines()
        # print(data)
    for i in data:
        if i.startswith('/'):
            continue
        main(i)
```
